```python
df = pd.DataFrame()
for i in ativos:
    frame = data.DataReader(i, data_source='yahoo', start='2015-01-01', end='2022-01-01')
    frame = frame['Adj Close']
    frame = pd.DataFrame(frame)
    frame = frame.rename(columns={'Adj Close':f'{i}'})
    df = pd.concat([frame, df], axis = 1)
```
